```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('src_imgs/01.png', cv2.IMREAD_COLOR)
img.shape
```

```python
img = cv2.imread('src_imgs/01_mask.png', cv2.IMREAD_GRAYSCALE)
```

```python
img.shape
```

```python
height, width = img.shape
c = np.zeros((height, width, 4))
for i in range(c.shape[2]):
    c[:,:,i] = (img == (i+1))

c = np.array([c])
```

```python
c.shape
```

```python
height, width = img.shape
c = np.zeros((1, height, width, 4))
c[0][:,:,0] = img == 1
c[0][:,:,1] = img == 2
c[0][:,:,2] = img == 3
c[0][:,:,3] = img == 4
```

```python
c.shape
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(25, 25))
axes[0].imshow(img)
axes[1].imshow(c[0][:,:,3])

plt.show()
```
